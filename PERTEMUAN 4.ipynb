```python
def kelipatan5():
    angka = int(input("Masukkan angka: "))
    if angka % 5 == 0:
        print("Kelipatan 5")
    else:
        print("Bukan kelipatan 5")

def tahun_kabisat():
    tahun = int(input("Masukkan tahun: "))
    if tahun % 400 == 0 or (tahun % 4 == 0 and tahun % 100 != 0):
        print("Tahun Kabisat")
    else:
        print("Bukan Tahun Kabisat")

def ganjil_genap():
    bil = int(input("Masukkan bilangan: "))
    if bil % 2 == 0:
        print("Bilangan Genap")
    else:
        print("Bilangan Ganjil")

def nama_bulan():
    bulan = int(input("Masukkan nomor bulan (1-12): "))
    nama_bulan = [
        "Januari", "Februari", "Maret", "April", "Mei", "Juni",
        "Juli", "Agustus", "September", "Oktober", "November", "Desember"
    ]
    if 1 <= bulan <= 12:
        print("Bulan:", nama_bulan[bulan - 1])
    else:
        print("Input salah")

def salam_jam():
    jam = int(input("Masukkan jam (0-23): "))
    if 0 <= jam <= 11:
        print("Selamat Pagi")
    elif 12 <= jam <= 15:
        print("Selamat Siang")
    elif 16 <= jam <= 18:
        print("Selamat Sore")
    elif 19 <= jam <= 23:
        print("Selamat Malam")
    else:
        print("Jam tidak valid")

while True:
    print("\n=== MENU ===")
    print("1. Cek Kelipatan 5")
    print("2. Tahun Kabisat")
    print("3. Ganjil / Genap")
    print("4. Nama Bulan")
    print("5. Salam Berdasarkan Jam")
    print("0. Keluar")
    pilih = input("Pilih menu: ")

    if pilih == "1":
        kelipatan5()
    elif pilih == "2":
        tahun_kabisat()
    elif pilih == "3":
        ganjil_genap()
    elif pilih == "4":
        nama_bulan()
    elif pilih == "5":
        salam_jam()
    elif pilih == "0":
        print("Terima kasih!")
        break
    else:
        print("Menu tidak valid.")
```

```python

print("="*30)
print("SELAMAT DATANG DI KAFE GEMINI")
print("="*30)
print("Menu Makanan & Minuman:")
print("1. Kopi Hitam  - Rp 12.000")
print("2. Teh Manis   - Rp  8.000")
print("3. Roti Bakar  - Rp 15.000")
print("4. Mie Instan  - Rp 10.000")
print("="*30)

pilihan_str = input("Silakan masukkan nomor pilihan Anda (1-4): ")
pilihan_int = int(pilihan_str)
nama_pesanan = ""
harga = 0

if pilihan_int == 1:
    nama_pesanan = "Kopi Hitam"
    harga = 12000
elif pilihan_int == 2:
    nama_pesanan = "Teh Manis"
    harga = 8000
elif pilihan_int == 3:
    nama_pesanan = "Roti Bakar"
    harga = 15000
elif pilihan_int == 4:
    nama_pesanan = "Mie Instan"
    harga = 10000
else:
    print("Maaf, pilihan menu tidak valid. Silakan jalankan ulang program.")
    exit()  # keluar program jika input tidak valid

print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga:,}")

print("\nKategori Diskon:")
print("1. Diskon 70%")
print("2. Diskon 50%")
print("3. Diskon 30%")
kategori = input("Pilih kategori diskon (1/2/3) atau tekan Enter jika tidak ada diskon: ")

diskon = 0
if kategori == "1":
    diskon = 0.70
elif kategori == "2":
    diskon = 0.50
elif kategori == "3":
    diskon = 0.30
else:
    diskon = 0  # tidak ada diskon

potongan = harga * diskon
total_bayar = harga - potongan
print(f"Potongan Diskon : Rp {potongan:,}")
print(f"Total Bayar     : Rp {total_bayar:,}")

uang_dibayar = int(input("\nMasukkan jumlah uang yang dibayar: Rp "))
if uang_dibayar >= total_bayar:
    kembalian = uang_dibayar - total_bayar
    print("================================")
    print("       STRUK PEMBAYARAN        ")
    print("Pesanan        :", nama_pesanan)
    print("Harga Awal     : Rp", format(harga, ","))
    print("Diskon         : Rp", format(potongan, ","))
    print("Total Bayar    : Rp", format(total_bayar, ","))
    print("Uang Dibayar   : Rp", format(uang_dibayar, ","))
    print("Kembalian      : Rp", format(kembalian, ","))
    print("================================")
    print("Terima kasih telah berbelanja!")
else:
    kekurangan = total_bayar - uang_dibayar
    print(f"Uang Anda kurang sebesar Rp {kekurangan:,}")
```
