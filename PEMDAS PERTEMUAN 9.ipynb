```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv")

df['quantity'].hist()
plt.title("Distribusi Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frekuensi")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("sales.csv")

# Korelasi
corr_matrix = df[['price_per_unit', 'quantity', 'total_price']].corr()
print("Korelasi:")
print(corr_matrix)

# Kovarian
cov_matrix = df[['price_per_unit', 'quantity', 'total_price']].cov()
print("\nKovarian:")
print(cov_matrix)
```
